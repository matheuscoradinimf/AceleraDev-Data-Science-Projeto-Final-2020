```python
import pandas as pd
import numpy as np
import pickle

from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import normalize
from sklearn.decomposition import NMF
from sklearn.model_selection import cross_val_score

from sklearn.model_selection import train_test_split
```

```python
csv_path = 'C:\\Data Science\\aceleradev\\Projeto_Final\\'

df = pd.read_csv(f'{csv_path}pp_market.csv')
port1 = pd.read_csv(f'{csv_path}estaticos_portfolio1.csv', index_col='Unnamed: 0')
port2 = pd.read_csv(f'{csv_path}estaticos_portfolio2.csv', index_col='Unnamed: 0')
port3 = pd.read_csv(f'{csv_path}estaticos_portfolio3.csv', index_col='Unnamed: 0')

ids = df['id']
df.drop('id',1,inplace=True)
```

```python
#scaler = StandardScaler()
scaler = MaxAbsScaler()

df=pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
#df.drop('Unnamed: 0',1,inplace=True)
df.index = ids
```

## Train test split

```python
port1_train, port1_test = train_test_split(port1, test_size = 0.3, random_state = 42)
port2_train, port2_test = train_test_split(port2, test_size = 0.3, random_state = 42)
port3_train, port3_test = train_test_split(port3, test_size = 0.3, random_state = 42)
```

## NMF

```python
def ranking(port, nlargest, df):
    port_train = port['id']
    teste = df.loc[port_train.values].T
    similarities = df.dot(teste)
    similarities = similarities.mean(axis=1)
    return similarities.nlargest(nlargest)
```

```python
port1_nmf = []
port2_nmf = []
port3_nmf = []

def hyperparam_nmf(n):
    nmf = NMF(n)
    nmf_features = nmf.fit_transform(df)
    norm_features = normalize(nmf_features)
    df_nmf = pd.DataFrame(norm_features, ids)
    
    for train,test,lista in [(port1_train,port1_test,port1_nmf),(port2_train,port2_test,port2_nmf),(port3_train,port3_test,port3_nmf)]:
        rank = ranking(train,20000,df_nmf).index
        count = 0
        for i in test.id:
            if i in rank:
                count = count + 1
        perc = count/len(test)
        print(f'Percentual: {perc}')
        lista.append(perc)
```

```python
n_nmf = []

for i in [5,10,15,20,25,30,35,40,45,50,55,60,65,70,75,80]:
    print(f'Número: {i}')
    hyperparam_nmf(i)
    n_nmf.append(i)
```

### SVD

```python
from sklearn.decomposition import TruncatedSVD
def ranking_svd(port, nlargest, df):
    port_train = port['id']
    teste = df.loc[port_train.values].T
    similarities = df.dot(teste)
    similarities = similarities.mean(axis=1)
    return similarities.nlargest(nlargest)
```

```python
port1_svd = []
port2_svd = []
port3_svd = []

def hyperparam_svd(n):
    svd = TruncatedSVD(n)
    nmf_features = svd.fit_transform(df)
    norm_features = normalize(nmf_features)
    df_svd = pd.DataFrame(norm_features, ids)
    
    for train,test,lista in [(port1_train,port1_test,port1_svd),(port2_train,port2_test,port2_svd),(port3_train,port3_test,port3_svd)]:
        rank = ranking(train,20000,df_svd).index
        count = 0
        for i in test.id:
            if i in rank:
                count = count + 1
        perc = count/len(test)
        print(f'Percentual: {perc}')
        lista.append(perc)
```

```python
n_svd = []

for i in [5,10,15,20,25,30,35,40,45,50,55,60,65,70,75,80]:
    print(f'Número: {i}')
    hyperparam_svd(i)
    n_svd.append(i)
```

```python
dictionary = {'port1_nmf': port1_nmf, 'port2_nmf':port2_nmf, 'port3_nmf':port3_nmf,
              'port1_svd':port1_svd,'port2_svd':port2_svd,'port3_svd':port3_svd,
             'n_nmf':n_nmf,'n_svd':n_svd}
```

```python
df_hyper = pd.DataFrame.from_dict(dictionary)
df_hyper.to_csv('hyperparam40000.csv')
```
