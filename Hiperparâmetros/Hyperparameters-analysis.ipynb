```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
h1000 = pd.read_csv('hyperparam1000.csv', index_col='Unnamed: 0')
h4000 = pd.read_csv('hyperparam4000.csv', index_col='Unnamed: 0')
h20000 = pd.read_csv('hyperparam20000.csv', index_col='Unnamed: 0')
h40000 = pd.read_csv('hyperparam40000.csv', index_col='Unnamed: 0')
```

```python
port = ['Portfólio 1','Portfólio 2','Portfólio 3']
```

```python
h1000.head(2)
```

```python
def plot_nmf(n):
    plt.figure(figsize=(18,7))
    plt.plot(n.n_nmf, 100*n.port1_nmf, color='darkturquoise')
    plt.plot(n.n_nmf, 100*n.port2_nmf, color='orchid')
    plt.plot(n.n_nmf, 100*n.port3_nmf, color='royalblue')
    plt.legend(port)
    plt.grid()
    plt.xlabel('Número de componentes', fontsize=20)
    plt.ylabel('Acerto (%)', fontsize=20)
    plt.title('NMF', fontsize=20)
    plt.show()
    
def plot_svd(n):
    plt.figure(figsize=(18,7))
    plt.plot(n.n_svd, 100*n.port1_svd, color='darkturquoise')
    plt.plot(n.n_svd, 100*n.port2_svd, color='orchid')
    plt.plot(n.n_svd, 100*n.port3_svd, color='royalblue')
    plt.legend(port)
    plt.grid()
    plt.xlabel('Número de componentes', fontsize=20)
    plt.ylabel('Acerto (%)', fontsize=20)
    plt.title('SVD', fontsize=20)
    plt.show()
```

# TOP 1000

```python
plot_nmf(h1000)
```

```python
plot_svd(h1000)
```

# TOP 4000

```python
plot_nmf(h4000)
```

```python
plot_svd(h4000)
```

# TOP 20000

```python
plot_nmf(h20000)
```

```python
plot_svd(h20000)
```

# TOP 40000

```python
plot_nmf(h40000)
```

```python
plot_svd(h40000)
```
